*Describe the probability space associated with the roulette wheel: Outcomes, events, probabilities (If there are 37 outcomes, there are $2^{37}= 137,438,953,472$ events, by the way)*
- Each discrete outcome has a probability of $\frac{1}{37}$. The sample space is the set of all possible outcomes {0,1,2,3,...,36}
- There are MANY events. The event space is the powerset of all possible outcomes. The more possible outcomes one chooses, the higher the probability. Events can be represented by numbers, or colors. For example, if a player bets on *black*, there is an $\frac{18}{37}$ probability of winning, because there are 18 of 37 possible outcomes of equal probability that are denoted with the black color.   


```python
# You wrote code to generate a spin of the roulette wheel. 
# Write a function that takes a basic or complex bet as an argument, and returns the result for the player (win or lose, and the payout 36/K-1 or -1)

def roulette_fxn(num_bet=None, color_bet = None, even_odd_bet= None):
    '''
    Choose one parameter at a time to pass to the function
    num_bets: pass as a list of integers between 0-36
    color_bets: pass as a list of color ie ["black"]
    even_odd_bet: pass as "even" or "odd"
    '''
    outcome = spin()
    # subset the DF based on the bet
    if num_bet:
        bet_list = roulette[roulette['Number'].isin(num_bet)]['Number'].tolist()
    elif color_bet:
        bet_list = roulette[roulette['Color'].str.lower().isin(color_bet)]['Number'].tolist()
    elif even_odd_bet:
        if even_odd_bet.lower() == 'even':
            bet_list = roulette[roulette['Number'] % 2 == 0]['Number'].tolist()
        elif even_odd_bet.lower() == 'odd':
            bet_list = roulette[roulette['Number'] % 2 != 0]['Number'].tolist()
    bet_size = len(bet_list)
    if outcome[0] in bet_list:
        payout = 36/bet_size-1
    else:
        payout = -1
    return payout
```

```python
# Betting on betting on red, betting on odd, betting on 7, a split, and a line 1000 times each.
# Compute the average values for the bets you just simulated. What are the expected average payoffs?
red_payout = []
seven_payout = []
split_payout = []
line_payout = []


for i in range(1000):
    red_payout.append(roulette_fxn(color_bet=['red']))  
    seven_payout.append(roulette_fxn(num_bet=[7]))  
    split_payout.append(roulette_fxn(num_bet=[1,2]))
    line_payout.append(roulette_fxn(num_bet=[1,2,3,4,5,6]))
```

```python
# Average Payouts:
avg_red = np.mean(red_payout)
avg_seven = np.mean(seven_payout)
avg_split = np.mean(split_payout)
avg_line = np.mean(line_payout)

print(f'The average Red Payout is: ${avg_red:,.2f}')
print(f'The average Seven Payout is: ${avg_seven:,.2f}')
print(f'The average Split Payout is: ${avg_split:,.2f}')
print(f'The average Line Payout is: ${avg_line:,.2f}')
```

Above are the expected payoffs after 1000 runs of the game, for different bets. On average the payoff is just below 0. When we simulate this game even more times the payoff tends to end up around $-.05, which is a an average loss of five cents.

$$
where $0 \le p_i \le 1$ and $ \sum_{k=1}^K p_k = 1$.

where $0 \le p_i \le 1$ and $ \sum_{k=1}^K p_k = 1$.


The **support** is {1,2,3...,K} for each discrete category. 

```python
# Plot the PMF / CDF
cat_grid = np.arange(1,6)
probabilities = np.array([.1,.2,.4,.2,.1])

cats = sp.stats.rv_discrete(name='categorical', values=(cat_grid, probabilities))

cdf = cats.cdf(cat_grid)
pmf = cats.pmf(cat_grid)

fig, ax = plt.subplots(1,2, figsize=(14,6))
ax[0].step(x=cat_grid, y=cdf)
ax[0].set_title('CDF') # Looks like a step function
sns.barplot(x=cat_grid, y=pmf, ax=ax[1])
ax[1].set_title('PMF')
plt.show()
```

```python
# Sample of 1000
sample = cats.rvs(size=1000, random_state=10)-1

fig, ax = plt.subplots(1,2, figsize=(14,6))
sns.ecdfplot(x=sample, ax=ax[0], label='ECDF')
ax[0].step(x=cat_grid, y=cdf, linestyle="--", color='orange', label='CDF')
ax[0].set_title('Sample CDF') # Looks like a step function
ax[0].legend()
sns.barplot(x=cat_grid, y=pmf, ax=ax[1], alpha=.3, label='PMF', color='orange')
sns.kdeplot(x=sample, ax=ax[1], label='KDE')
ax[1].set_title('Sample PMF')
ax[1].legend()
plt.show()
```
